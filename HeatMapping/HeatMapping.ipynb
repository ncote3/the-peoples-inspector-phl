Heat Mapping of Various 311 Data

First need to grab the imports

```python
import pandas as pd
import plotly.graph_objects as go
```

Then we need to put the data in a pandas data frame

We should describe the typing of the data for memory efficiency 

```python
requests = pd.read_csv('../data/311_requests.csv', low_memory=False).convert_dtypes()  
```

```python
fig = go.Figure(go.Densitymapbox(lat=requests.lat, lon=requests.lon, radius=3))
fig.update_layout(mapbox_style="stamen-terrain", mapbox_center_lon=180)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
illegal_dumping_requests = requests[requests['service_name'] == "Illegal Dumping"].head(100000)
```

```python

illegal_dumping_requests_fig = go.Figure(go.Densitymapbox(lat=illegal_dumping_requests.lat, lon=illegal_dumping_requests.lon, radius=1))
illegal_dumping_requests_fig.update_layout(mapbox_style="stamen-terrain", mapbox_center_lon=180)
illegal_dumping_requests_fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
illegal_dumping_requests_fig.show()
```
